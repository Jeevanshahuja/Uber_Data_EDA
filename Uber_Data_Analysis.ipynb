## Doing the necessary data cleaning steps like dropping duplicates, unnecessary columns, null value imputation, outliers treatment etc

```python
def convert_to_datetime(date_str):
    formats = ['%d-%m-%Y %H:%M', '%d/%m/%Y %H:%M']
    for fmt in formats:
        try:
            return pd.to_datetime(date_str, format=fmt)
        except ValueError:
            pass
    return pd.NaT
df['PICKUP DATE'] = df['PICKUP DATE'].apply(convert_to_datetime)
```

## REMOVING OUTLIERS

```python
for numerical in ["PICKUPS","WIND SPEED","VISIBILITY","TEMPERATURE"]:
    plt.figure(figsize=(10, 5))
    Q1 = df[numerical].quantile(0.25)
    Q3 = df[numerical].quantile(0.75)
    IQR = Q3 - Q1
    filtered_data = df[(df[numerical] >= (Q1 - 1.5 * IQR)) & (df[numerical] <= (Q3 + 1.5 * IQR))][numerical]
    sns.boxplot(data=filtered_data)
    plt.title(f'Box plot of {numerical} without outliers')
    plt.show()
```

```python
def remove_outliers_iqr(df, numeric_cols):
    df_filtered = df.copy()
    for col in numeric_cols:
        Q1 = df_filtered[col].quantile(0.25)
        Q3 = df_filtered[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df_filtered = df_filtered[(df_filtered[col] >= lower_bound) & (df_filtered[col] <= upper_bound)]
    
    return df_filtered
# Remove outliers
df_filtered = remove_outliers_iqr(df,["PICKUPS","WIND SPEED","VISIBILITY","TEMPERATURE"] )
print("Original DataFrame shape:", df.shape)
print("DataFrame without outliers shape:", df_filtered.shape)
```

```python
sns.boxplot(data=df_filtered,orient='h')
```

## Null value imputation

```python
# Impute null values for numerical columns with the mean
df_filtered[numeric] = df_filtered[numeric].fillna(df_filtered[numeric].median())
# Impute null values for categorical columns with the mode
df_filtered[categorical] = df_filtered[categorical].fillna(df_filtered[categorical].mode().iloc[0])
# Display the updated DataFrame
print(df_filtered.head())
```

## Number of pickups compare between weekdays and weekends for each borough

```python
weekday_weekend_pickup_counts = df_filtered.groupby(['BOROUGH', df_filtered['PICKUP DATE'].dt.dayofweek])['PICKUPS'].sum().reset_index()
weekday_weekend_pickup_counts.columns = ["BOROUGH", "DAY", "PICKUPS"]
boroughs = weekday_weekend_pickup_counts['BOROUGH'].unique()
weekdays = []
weekends = []
for borough in boroughs:
    borough_data = weekday_weekend_pickup_counts[weekday_weekend_pickup_counts['BOROUGH'] == borough]
    weekday_data = borough_data[borough_data['DAY'].isin([0, 1, 2, 3, 4])]
    weekend_data = borough_data[borough_data['DAY'].isin([5, 6])]
    weekdays.append(weekday_data['PICKUPS'].sum())
    weekends.append(weekend_data['PICKUPS'].sum())
pickup_sums_df = pd.DataFrame({
    'Borough': boroughs,
    'Weekday Pickups': weekdays,
    'Weekend Pickups': weekends
})
fig, ax = plt.subplots(figsize=(14, 8))
bar_width = 0.35
index = range(len(boroughs))
bar1 = ax.bar(index, pickup_sums_df['Weekday Pickups'], bar_width, label='Weekday Pickups', edgecolor="black", linewidth=3)
bar2 = ax.bar([i + bar_width for i in index], pickup_sums_df['Weekend Pickups'], bar_width, label='Weekend Pickups', edgecolor="black", linewidth=3)
ax.set_xlabel('Borough', fontsize=14)
ax.set_ylabel('Number of Pickups', fontsize=14)
ax.set_title('Weekday and Weekend Uber Pickups by Borough', fontsize=16)
ax.set_xticks([i + bar_width / 2 for i in index])
ax.set_xticklabels(boroughs)
ax.legend()
plt.show()
```

```python
df_filtered['MONTH']=df_filtered['PICKUP DATE'].dt.month
```

```python
def month_to_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Fall'
df_filtered['SEASON'] = df_filtered['MONTH'].apply(month_to_season)
```

```python
season_pickup=df_filtered.groupby("SEASON")["PICKUPS"].sum()
season_pickup
```
